```python
def crawling_board(name, rank, tgd_link):  
    df_stream_board = pd.DataFrame(columns=li_stream_article)
    
    board_url = "https://tgd.kr" + tgd_link
    response_baord = requests.get(board_url, headers = {"cookie" : cookies})
    dom_board = BeautifulSoup(response_baord.content, "html.parser")
    
    #get last page
    try:
        li_page = dom_board.select("#article-list-menu li a")
        last_page = int(li_page[len(li_page)-1]["data-ci-pagination-page"])

        if last_page > 50:
            last_page = 50

        for page in tqdm_notebook(range(1, last_page+1), desc="page_num", leave=False):
            # get article link in page
            page_url = "https://tgd.kr" + tgd_link + "/page/" + str(page)
            response_page = requests.get(page_url, headers = {"cookie" : cookies})
            dom_page = BeautifulSoup(response_page.content, "html.parser")

            li_article = dom_page.select(".article-list-row")

            li_article_link = []
            try:
                for article in li_article:
                    if "notice" not in article["class"]:
                        li_article_link.append(article.select_one("a").get('href'))
            except AttributeError as e:

                pass

            # get articles
            for article_link in tqdm_notebook(li_article_link, desc="article_num", leave=False):
                s_stream_board = pd.Series(index = li_stream_article)

                article_url = "https://tgd.kr" + article_link
                response_article = requests.get(article_url, headers = {"cookie" : cookies})
                dom_article = BeautifulSoup(response_article.content, "html.parser")

                # get article info
                try:
                    s_stream_board["title"] = dom_article.select_one("#article-info h2").text.split("\n")[-1].strip()
                    s_stream_board["writer"] = dom_article.select_one("#article-info-writer strong").text
                    s_stream_board["writerTime"] = dom_article.select_one("#article-time span").get("title")
                    s_stream_board["commentNum"] = dom_article.select_one("#article-comment").text.strip()
                    s_stream_board["likeNum"] = dom_article.select_one("#article-votes").text.strip()
                    s_stream_board["hateNum"] = dom_article.select_one("#down").text
                    s_stream_board["viewNum"] = dom_article.select_one("#article-views").text.strip()

                    s_stream_board["contents"] = dom_article.select_one("#article-content").text

                # get comment info
                    comment_url = "https://tgd.kr/board/comment_load"+ article_link
                    comments_json = requests.get(comment_url).json()

                    li_comment_nickname = []
                    comment_content = ""

                    for comment in comments_json["data"]:
                        li_comment_nickname.append(comment["nickname"])
                        comment_content += (BeautifulSoup(comment["content"], "html.parser").text + " ")

                    s_stream_board["commentUniqueNum"] = len(set(li_comment_nickname))
                    s_stream_board["commentContent"] = re.sub("\s", " ",comment_content)
                except AttributeError as e:
                    pass

                df_stream_board.loc[len(df_stream_board)] = s_stream_board

        path_save = "data/tgd/" + str(rank) + "_" + str(name) + '.csv'      
        df_stream_board.to_csv(path_save, index = False)
    except Exception as e:
            print("로그인이 필요 !! {}".format(name))
```
